## Tasks

Now write and execute SQL queries to solve the assignment tasks.

**Note: If the column names are in mixed case enclose it in double quotes
   For Example "Landing_Outcome"**

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
query = "SELECT DISTINCT(Launch_Site) FROM SPACEXTBL"
task1 = pd.read_sql(query, con)
task1
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
query = "SELECT * FROM SPACEXTBL where Launch_Site like 'CCA%' limit 5"
task2 = pd.read_sql(query, con)
task2
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
query = "SELECT 'Nasa (CRS)', sum(PAYLOAD_MASS__KG_) FROM SPACEXTBL where Customer like '%CRS%' group by 1"
task3 = pd.read_sql(query, con)
task3
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
query = "SELECT 'Booster F9 v1.1' as Booster_version, avg(PAYLOAD_MASS__KG_) as avg_payload FROM SPACEXTBL where Booster_Version like '%F9 v1.1%' group by 1"
task4 = pd.read_sql(query, con)
task4
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
query = "SELECT distinct(Landing_Outcome) from SPACEXTBL"
task5 = pd.read_sql(query, con)
task5
```

```python
query = "SELECT min(Date) from SPACEXTBL where Landing_Outcome = 'Success (ground pad)'"
task5 = pd.read_sql(query, con)
task5
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
task6 = pd.read_sql("""
            SELECT 
                Booster_Version,  
                avg(PAYLOAD_MASS__KG_) as avg_payload
            from 
                SPACEXTBL 
            where 
                Landing_Outcome = 'Success (drone ship)' 
                and PAYLOAD_MASS__KG_ between 4000 and 6000
            group by
                1""", con)
task6
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
task7 = pd.read_sql("""
            SELECT 
                Mission_Outcome,  
                count(Date) as Total
            from 
                SPACEXTBL
            group by 1""", con)
task7
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
task8 = pd.read_sql("""
            SELECT 
                Booster_Version,
                max(PAYLOAD_MASS__KG_) payload
            from 
                SPACEXTBL
            where
                PAYLOAD_MASS__KG_ = (
                                    select 
                                        max(PAYLOAD_MASS__KG_)
                                    from SPACEXTBL                                  
                                    )
            
            """, con)
task8
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 4, 2) as month to get the months and substr(Date,7,4)='2015' for year.**


```python
task9 = pd.read_sql("""
            SELECT 
                substr(Date,4,2) as Month,
                Landing_Outcome,
                Booster_Version,
                Launch_Site
            from 
                SPACEXTBL
            where
                Landing_Outcome = 'Failure (drone ship)'
                and substr(Date,7,4)='2015'
            """, con)
task9
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
task10 = pd.read_sql("""
            SELECT 
                Landing_Outcome,
                count(Date) as Total
            from 
                SPACEXTBL
            where
                Date between '04/06/2010' and '20/03/2017'
            group by 1
            order by 2 desc
            """, con)
task10
```
